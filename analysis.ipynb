```python
import pandas as pd
from numpy import mean
#相关系数计算公式
def R_Square(x,y):
    p1=x2=y2=0.0
    #计算平均值
    x_=mean(x)
    y_=mean(y)
    #循环读取每个值，计算对应值的累和
    for i in range(len(x)):
        p1+=(x[i]-x_)*(y[i]-y_)
        x2+=(x[i]-x_)**2
        y2+=(y[i]-y_)**2
    #print(p1,x2,y2)
    #计算相关系数
    r=p1/((x2** 0.5)*(y2** 0.5))
    return r
#读取数据
matrix_hour = np.zeros((357,357))
for i in range(357):
    for j in range(357):
        matrix_hour[i][j] = R_Square(nn_data[i][:],nn_data[j][:])
        
#调用并输出相关系数


```
